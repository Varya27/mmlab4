```python
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#№1
w=4.70
g=0.00
#№2
w=7
g=0.5
#№3
w=0.5
g=7
```

```python
def func(t):
    #f=0
    f=0.5*math.sin(0.7*t)
    return f
```

```python
x0=np.array([0.9,1.9])
```

```python
t=np.arange(0,56,0.05)
```

```python
def syst (x,t):
    dx_1=x[1]
    dx_2=-w*w*x[0]-g*x[1]-func(t)
    return [dx_1, dx_2]
```

```python
x=odeint(syst, x0, t)
```

```python
plt.plot([i[0] for i in x], [i[1] for i in x], lw=2)
plt.title('Задание №3')
plt.grid(True)
```
